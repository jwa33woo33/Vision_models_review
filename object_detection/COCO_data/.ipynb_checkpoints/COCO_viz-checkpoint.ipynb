```python
import sys
import json
import os
import random
import argparse
import math
import tqdm
from collections import defaultdict
from typing import Any, Callable, TypeVar

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from matplotlib.patches import Polygon, Rectangle
from matplotlib.collections import PatchCollection

import pycocotools
from pycocotools import mask
```

```python
# Load the dataset json
class CocoDataset():
    def __init__(self, annotation_path, image_dir, image_file):
        self.annotation_path = annotation_path
        self.image_dir = image_dir
        self.image_file = image_file

        json_file = open(self.annotation_path)
        self.coco = json.load(json_file)
        json_file.close()

        self.process_categories()


    def display_categories(self):
        print('Categories:')
        print('=========')
        for sc_key, sc_val in self.super_categories.items():
            print('  super_category: {}'.format(sc_key))
            for cat_id in sc_val:
                self.category_name = self.categories[cat_id]['name']
                print(f'    id {cat_id}: {self.category_name}')
            print('')


    def process_categories(self):
        self.categories = {}
        self.super_categories = {}
        for category in self.coco['categories']:
            cat_id = category['id']
            super_category = category['supercategory']

            # Add category to the categories dict
            if cat_id not in self.categories:
                self.categories[cat_id] = category
            else:
                print("ERROR: Skipping duplicate category id: {}".format(category))

            # Add category to super_categories dict
            if super_category not in self.super_categories:
                # Create a new set with the category id
                self.super_categories[super_category] = {cat_id}
            else:
                self.super_categories[super_category] |= {
                    cat_id}  # Add category id to the set
                
    def visualize(self):
        with open(annotation_path) as json_file:
            data = json.load(json_file)
        
        fig, (ax1, ax2) = plt.subplots(1, 2, dpi=300)
        image = Image.open(self.image_dir+self.image_file)
        ax1.imshow(image)
        ax2.imshow(image)

        color = rgb_to_rgba((0, 128, 0), 0.5)

        for annot in data['annotations']:
            if annot['image_id'] == 6954:
                poly = np.asarray(annot['segmentation'])
                box = np.asarray(annot['bbox'])
                rectangle = Rectangle((box[0], box[1]), box[2], box[3], fc=color[0], ec=color[1])
                ax1.add_patch(rectangle)

                ax1.annotate(
                    coco_dataset.categories[annot['category_id']]['name'],
                    (box[0], box[1]), color='w', weight='bold', 
                    fontsize=4, ha='left', va='bottom', 
                    bbox=dict(facecolor=color[1], edgecolor=color[1], pad=0.0))

                poly = poly.reshape(-1, 2)
                polygon = Polygon(poly, True, fc=color[0], ec=color[1], lw=0.5)
                ax2.add_patch(polygon)

                ax2.annotate(
                    coco_dataset.categories[annot['category_id']]['name'],
                    (box[0], box[1]), color='w', weight='bold', 
                    fontsize=4, ha='left', va='bottom', 
                    bbox=dict(facecolor=color[1], edgecolor=color[1], pad=0.0))


        ax1.axis('off')
        ax2.axis('off')
```

```python
annotation_path = './data/instances_val2014.json'
image_dir = './data/'
image_file = '000000006954.jpg'

def rgb_to_rgba(rgb, alpha = 0.3):
    rgb = np.asarray(rgb) / 255.
    c1 = np.insert(rgb, len(rgb), alpha).round(2)
    c2 = np.insert(rgb, len(rgb), 1.).round(2)
    return (tuple(c1), tuple(c2))
```

```python
coco_dataset = CocoDataset(annotation_path, image_dir, image_file)
```

```python
coco_dataset.visualize()
```
